```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model


# Load the data
oecd_bli = pd.read_csv("oecd_bli_2015.csv", thousands=',')
gdp_per_capita = pd.read_csv("gdp_per_capita.csv", thousands=',', delimiter='\t', encoding='latin1', na_values="n/a")

# Prepare Country Stats Function
def prepare_country_stats(oecd_bli, gdp_per_capita):
    oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
    oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
    gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
    gdp_per_capita.set_index("Country", inplace=True)
    full_country_stats = pd.merge(left=oecd_bli, right=gdp_per_capita,
                                  left_index=True, right_index=True)
    full_country_stats.sort_values(by="GDP per capita", inplace=True)
    return full_country_stats[["GDP per capita", "Life satisfaction"]]

# Prepare the data
country_stats = prepare_country_stats(oecd_bli, gdp_per_capita)
X = np.c_[country_stats["GDP per capita"]]
Y = np.c_[country_stats["Life satisfaction"]]

# Visualize the data
country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.show()

# Select a linear model
model = sklearn.linear_model.LinearRegression()

# Train the model
model.fit(X, Y)

# Make a prediction for India
X_new = [[41973]]  # India's GDP per capita
print(model.predict(X_new))
```

```python
# Predict House Prices Based on Area
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data
data = pd.read_csv("house_data.csv")
X = data[["Area"]]
Y = data[["Price"]]

# Visualize the data
plt.scatter(X,Y, color="green")
plt.xlabel("Area (sq. ft)")
plt.ylabel("Price ($)")
plt.title("House Price Prediction Model")
plt.show()

# Train the model
model = LinearRegression()
model.fit(X,Y)

# Predict Price of a house with 2500 sq. ft.
area_to_predict = [[2500]] # input data here
predicted_price = model.predict(area_to_predict)

print("Predicted Price for ",area_to_predict,"sq. ft. house: $",predicted_price)
```
